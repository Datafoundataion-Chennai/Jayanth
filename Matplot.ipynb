```python
import matplotlib.pyplot as plt
import numpy as np
days = np.arange(1, 11)  
stock_price = [150, 152, 149, 155, 160, 162, 158, 165, 170, 175] 
trading_volume = [1.2, 1.4, 1.1, 1.8, 2.0, 2.5, 2.2, 2.7, 3.0, 3.5]
#plt.plot(days,stock_price,linestyle = 'dotted',color='r')

fig, ax1 = plt.subplots()
ax1.set_xlabel('Days') 
ax1.set_ylabel('Stock_price', color = 'r')
ax1.plot(days, stock_price, color = 'r',linestyle="dashed") 

ax2 = ax1.twinx() 
ax2.set_ylabel('Trading_volume', color = 'g')
ax2.plot(days, trading_volume, color = 'g',linestyle="dashed")
fig.suptitle('Stock Price & Trade Volume', fontweight ="bold")
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
rent = [1200, 1200, 1200, 1200, 1250, 1250]
food = [400, 420, 450, 480, 500, 520]
entertainment = [150, 200, 220, 180, 250, 300]
transport = [100, 120, 130, 110, 140, 160]
plt.bar(months, rent, color='b')
plt.bar(months, food, color='g')
plt.bar(months, entertainment, color='y')
plt.bar(months, transport, color='r')
plt.xlabel("Months")
plt.ylabel("Expenditure")
plt.title("Monthly Expenses Comparison")
plt.legend(['Rent','Food','Entertainment','Transport'],loc='center right', bbox_to_anchor=(1.35, 0.5))


plt.show()
```

```python
weeks = np.arange(1, 9)  
covid_cases = [500, 700, 1200, 2000, 3000, 4500, 6000, 8000]  
vaccination_rate = [5, 10, 20, 35, 50, 65, 75, 85]  
plt.subplot(2, 1, 1)
plt.plot(weeks,covid_cases)
plt.legend(["covid_cases"])
plt.grid()
plt.title("Covid Cases")

plt.subplot(2, 1, 2)
plt.bar(weeks,vaccination_rate,color='r',linestyle="dashed")
plt.title("Vacination Rate")
plt.legend(["Vacination_Rate"])
plt.suptitle("COVID-19 Cases & Vaccination Progress")

plt.show()
```
